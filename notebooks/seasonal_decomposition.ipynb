```python
import hvplot.polars  # noqa
import polars as pl

from polars_ts.decomposition.seasonal_decomposition import seasonal_decomposition

# Load sample df
df = (
    pl.scan_parquet("https://datasets-nixtla.s3.amazonaws.com/m4-hourly.parquet")
    .filter(pl.col("unique_id") == "H1")
    .collect()
)
```

```python
freq = 24
# Call the seasonal_decomposition function
decomposed_df = seasonal_decomposition(
    df,
    freq=freq,  # specify this as the frequency of your data, so hourly frequency with daily seasonality....
    method="additive",  # You can also try 'multiplicative'
)
```

```python
decomposed_df.hvplot.line(x="ds", y=["y", "trend", f"seasonal"])
```
